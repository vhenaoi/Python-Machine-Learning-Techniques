# **Tuples**
tuples are **inmutable** lists

```python
# Initialize the variable `a` with a tuple containing four elements:
# - 10
# - 3
# - 4
# - 5
# Tuples are immutable sequences in Python, meaning their elements cannot be changed after creation.
a = (10, 3, 4, 5)
```

```python
# Access and display multiple aspects of the tuple `a`:
# - `a[3]` retrieves the element at index 3 of the tuple, which is the last element (5).
# - `a[::-1]` reverses the tuple, producing a new tuple with the elements in reverse order: (5, 4, 3, 10).
# - `sum(a)` calculates the sum of all elements in the tuple. Since `a` contains (10, 3, 4, 5), the sum is 22.
# Output: (5, (5, 4, 3, 10), 22)
a[3], a[::-1], sum(a)
```

# **Functions**
Recall that Python is an indented language. It does not uses any kind of brakets to delimit the scope of programming structures, which are signaled by **changes in indentation**

Arguments types are not declared, implementation must be prepared to deal with the data types you want

```python
# Define a function `f1` that takes two parameters `a` and `b`.
# - `r` is calculated as `a` multiplied by 2.
# - The function returns the sum of `r` and `b`.
# Example:
# - If `a` is 3 and `b` is 4, `r` will be 6, and the function will return 10 (6 + 4).

def f1(a, b):
    r = a * 2
    return r + b

# Define a function `gt` that takes two parameters `a` and `b`.
# - The function returns the greater of the two values `a` and `b`.
# Example:
# - If `a` is 5 and `b` is 3, the function will return 5.

def gt(a, b):
    if a > b:
        return a
    else:
        return b
```

```python
# Call the function `f1` with arguments `2` and `3`.
# - Inside `f1`, `r` is calculated as `2 * 2`, which is 4.
# - The function returns `r + 3`, which is `4 + 3`, resulting in 7.
# Output: 7
f1(2, 3)
```

```python
# Call the function `f1` with arguments `['hola', 2]` and `[3]`.
# - The function attempts to perform the operation `a * 2`, where `a` is `['hola', 2]`.
# - Multiplying a list by an integer replicates the list, so `['hola', 2] * 2` results in `['hola', 2, 'hola', 2]`.
# - The function then attempts to return the result of adding this list to `[3]`.
# - Adding a list to another list concatenates them, but Python will raise a `TypeError` because lists cannot be concatenated with lists of different types (e.g., adding a list to a list of integers).

# Output: TypeError: can only concatenate list (not "int") to list
f1(['Hello', 2], [3])
```

```python
# Call the function `f1` with arguments `'pepe'` and `'peppa'`.
# - Inside `f1`, `r` is calculated as `'pepe' * 2`, which results in the string `'pepepepe'` (the string `'pepe'` repeated twice).
# - The function then returns the result of concatenating `'pepepepe'` with `'peppa'`, resulting in the string `'pepepepepeppa'`.
# Output: 'pepepepepeppa'
f1('pepe', 'peppa')
```

```python
# Call the function `gt` with arguments `3` and `4`.
# - The function compares `3` and `4` using an if-else statement.
# - Since `3` is not greater than `4`, the function will return `4`, which is the value of `b`.
# Output: 4
gt(3, 4)
```

```python
# Call the function `gt` with arguments `"Hello"` and `"Bye"`.
# - The function compares the strings `"Hello"` and `"Bye"` using an if-else statement.
# - In Python, string comparison is based on lexicographic (alphabetical) order.
# - Since `"Bye"` is lexicographically less than `"Hello"`, the function will return `"Hello"`, which is the value of `b`.
# - In the ASCII or Unicode order:
# - `'B'` comes before `'H'`, meaning `'B'` has a lower value than `'H'`.
# Output: "Hello"
gt("Hello", "Bye")
```

functions can have default values

```python
# Define a function `f_power` that calculates the power of a number.
# - The function takes two parameters: `x` (the base) and `p` (the exponent), with `p` defaulting to 2.
# - It returns `x` raised to the power of `p` (i.e., `x**p`).
# Call the function `f_power` with the argument `3`.
# - Since no value is provided for `p`, the default value of 2 is used.
# - The function computes `3**2`, which equals `9`.
# Output: 9
def f_power(x, p=2):
    return x**p

f_power(3)
```

```python
# Call the function `f_power` with arguments `3` and `2`.
# - `x` is set to `3` and `p` is explicitly set to `2`.
# - The function calculates `3**2`, which is `9`, and returns this value.
# Output: 9
f_power(3, 2)
```

```python
# Call the function `f_power` with keyword arguments `p=2` and `x=3`.
# - In this case, `x` is set to `3` and `p` is set to `2`.
# - The function calculates `3**2`, which is `9`, and returns this value.
# Output: 9
f_power(p=2, x=3)
```

functions can have functions defined within, but they are only visible inside the function

```python
# Define a function `f` that takes two parameters: `a` and `b`.
# Inside `f`, define a nested function `squared` that takes one parameter `x`.
# - The `squared` function returns the square of `x` (i.e., `x**2`).
# The outer function `f` calculates the sum of `a` and `b`, and then calls the `squared` function with this sum as the argument.
# - `f` returns the result of squaring the sum of `a` and `b`.
# Example:
# - If `a` is 2 and `b` is 3, the sum is 5. The `squared` function will return `5**2`, which is 25.
# Output: 25 (for the example values)
def f(a,b):
    def squared(x):
        return x**2
    return squared(a+b)
```

```python
# Call the function `f` with arguments `2` and `3`.
# - Inside `f`, calculate the sum of `2` and `3`, which is `5`.
# - Call the nested function `squared` with `5` as the argument.
# - The `squared` function returns `5**2`, which is `25`.
# - The outer function `f` returns this result.
# Output: 25
f(2, 3)
```

functions are regular python objects which can be passed around

```python
# Define a function `apply_if_int` that takes two parameters: a function `f` and a value `x`.
# - The function checks if `x` is of type `int`.
# - If `x` is an integer, it applies the function `f` to `x` and returns the result.
# - If `x` is not an integer, it simply returns `x` without applying the function.
# Example:
# - If `x` is 5 and `f` is a function that squares a number, `apply_if_int` will return `f(5)` (which is 25).
# - If `x` is "hello", the function will return "hello" without modification.
def apply_if_int(f, x):
    if type(x)==int:
        return f(x)

    return x
```

```python
# Define a function `cubed` that takes one parameter `x`.
# - The function returns `x` raised to the power of 3 (i.e., `x**3`).
# Example:
# - If `x` is 2, the function will return `2**3`, which is 8.
def cubed(x):
    return x**3
```

```python
# Call the function `apply_if_int` with the `cubed` function and the integer `3` as arguments.
# - `apply_if_int` checks if `3` is of type `int`, which it is.
# - Since `3` is an integer, `apply_if_int` applies the `cubed` function to `3`.
# - The `cubed` function calculates `3**3`, which is `27`, and returns this value.
# Output: 27
apply_if_int(cubed, 3)
```

```python
# Call the function `apply_if_int` with the `cubed` function and the float `3.0` as arguments.
# - `apply_if_int` checks if `3.0` is of type `int`.
# - Since `3.0` is a float (not an integer), `apply_if_int` does not apply the `cubed` function.
# - The function simply returns the original value `3.0`.
# Output: 3.0
apply_if_int(cubed, 3.)
```

# **Condition**

```python
# Define a variable `a` with the value `15`.
# Check if `a` is greater than `10`.
# - Since `15` is greater than `10`, the condition is true.
# - Assign the string `"greater than 10"` to the variable `s`.
# If the condition were false, the string `"less than 10"` would be assigned to `s`.
# Print the value of `s`.
# Output: "greater than 10"
a=15
if a > 10:
    s = "greater than 10"
else:
    s = "less than 10"

print(s)
```

```python
# Define a variable `a` with the value `5`.
# Use a ternary conditional expression to assign a value to `s`.
# - The expression checks if `a` is greater than `10`.
# - Since `5` is not greater than `10`, the condition is false.
# - Therefore, `"less than 10"` is assigned to `s`.
# Print the value of `s`.
# Output: "less than 10"
a = 5
s = "greater than 10" if a > 10 else "less than 10"
print(s)
```

# **DataFrame indexing**
is **NOT** exactly like numpy

* first index
  * if string refers to columns
  * if Series of booleans is used as a filter
* for selecting columns:
  * use .loc to select by Index
  * use .iloc to select by position

```python
# Access the 'mmse' column using bracket notation
print(df_loaded["mmse"])  # Access and print the 'mmse' column from the DataFrame using bracket notation, which is always valid
```

```python
# Access the first row of the DataFrame using iloc
print(df_loaded.loc[0])  # Access and print the first row of the DataFrame using label-based indexing with loc. This will work if 0 is a valid index label.
```

Index semantics is exact!!

```python
# Use loc to select rows by index labels
'''Label-Based Indexing (loc): The loc method selects rows based on index
labels, and it includes the end label in the selection.'''
print(df_loaded.loc[0:7])  # Select and print rows from index label 0 to 7 (inclusive) using loc. This method is label-based, so it includes the end label.
```

```python
# Use iloc to select rows by integer positions
'''Position-Based Indexing (iloc): The iloc method selects rows based
on integer positions, and it excludes the end position.'''
print(df_loaded.iloc[10:15])  # Select and print rows from integer position 10 to 14 (exclusive of 15) using iloc. This method is position-based.
```
